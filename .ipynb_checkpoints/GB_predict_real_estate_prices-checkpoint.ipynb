## Feature Engineering

```python
# Create indicator variable for properties with 2 beds and 2 baths:
df['two_and_two'] = ((df.beds == 2) & (df.baths == 2)).astype(int)
# Create indicator variable for properties with 3 beds and 3 baths:
df['three_and_three'] = ((df.beds == 3) & (df.baths == 3)).astype(int)
# Create indicator variable for properties with 4 beds and 4 baths:
df['four_and_four'] = ((df.beds == 4) & (df.baths == 4)).astype(int)
```

```python
# Create a property age feature
df['property_age'] = df.year - df.yrbuilt
# Number of observations with 'property_age' < 0
print( sum(df.property_age < 0) )
# Remove rows where property_age is less than 0
df = df[df.property_age >= 0]
```

```python
#Define new categorical variables. Baths with levels 1,2,3,4+
df.baths.replace([6,5],4, inplace=True)
#Beds with levels 1,2,3,4,5+
df.beds.replace(6,5,inplace=True)
```

```python
#Define new categorical variables:
df=pd.get_dummies( df, columns=['zip','year','month',"baths","beds",'prevyear','prevmonth'])
df.head()
```

## Model training

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

from sklearn.model_selection import train_test_split

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

import pickle

import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.simplefilter(action='ignore', category=ConvergenceWarning)
```

```python
# Create separate object for target variable
y = df.price

# Create separate object for input features
X = df.drop('price', axis=1)
```

```python
# Split X and y into train, test and dev sets
X_train, X_dev, y_train, y_dev = train_test_split(X, y,
                                                   test_size=0.25,
                                                   random_state=123)

X_dev, X_test, y_dev, y_test = train_test_split(X_dev, y_dev,
                                                   test_size=0.5,
                                                   random_state=123)
```

```python
print( '{} observations in the training set.'.format( len(X_train) ) )
print( '{} observations in the dev set.'.format( len(X_dev) ) )
print( '{} observations in the test set.'.format( len(X_test) ) )
```

```python
# Create pipelines dictionary for lasso and ridge regressions, elastic net, random forest and gradient boosting:
pipelines = {
    'lasso' : make_pipeline(StandardScaler(), Lasso(random_state=123)),
    'ridge' : make_pipeline(StandardScaler(), Ridge(random_state=123)),
    'enet'  : make_pipeline(StandardScaler(), ElasticNet(random_state=123)),
    'rf'    : make_pipeline(StandardScaler(), RandomForestRegressor(random_state=123)),
    'gb'    : make_pipeline(StandardScaler(), GradientBoostingRegressor(random_state=123))
}

# Lasso hyperparameters
lasso_hyperparameters = { 
    'lasso__alpha' : [0.0001, 0.001, 0.01, 0.1, 1, 5, 10] 
}

# Ridge hyperparameters
ridge_hyperparameters = { 
    'ridge__alpha': [0.0001, 0.001, 0.01, 0.1, 1, 5, 10]  
}

# Elastic Net hyperparameters
enet_hyperparameters = { 
    'elasticnet__alpha': [0.0001, 0.001, 0.01, 0.1, 1, 5, 10],                        
    'elasticnet__l1_ratio' : [0.1, 0.3, 0.5, 0.7, 0.9]  
}

# Random forest hyperparameters
rf_hyperparameters = { 
    'randomforestregressor__n_estimators' : range(100,701,200),
    'randomforestregressor__max_features': ['auto', 'sqrt', 0.33]
}

# Boosted tree hyperparameters
gb_hyperparameters = { 
    'gradientboostingregressor__n_estimators': range(100,901,300),
    'gradientboostingregressor__learning_rate' : [0.05, 0.1, 0.15],
    'gradientboostingregressor__max_depth': [3, 5, 7]
}
```

```python
# Create hyperparameters dictionary
hyperparameters = {
    'rf' : rf_hyperparameters,
    'gb' : gb_hyperparameters,
    'lasso' : lasso_hyperparameters,
    'ridge' : ridge_hyperparameters,
    'enet' : enet_hyperparameters
}
```

```python
# Create empty dictionary called fitted_models
fitted_models = {}

# Loop through model pipelines, tuning each one and saving it to fitted_models
for name, pipeline in pipelines.items():
    
    model = GridSearchCV(pipeline, hyperparameters[name], cv = 3, verbose=0, n_jobs = -1)
    
    # Fit model on X_train, y_train
    model.fit(X_train, y_train)
    
    # Store model in fitted_models[name] 
    fitted_models[name] = model
    
    # Print '{name} has been fitted'
    print(name, 'has been fitted.')
```

```python
for name, model in fitted_models.items():
    print( name, model.best_score_ )
```

```python
#Preformance on the train set:
for name, model in fitted_models.items():
    pred = model.predict(X_train)
    print( name )
    print( '--------' )
    print( 'R^2:', r2_score(y_train, pred ))
    print( 'MAE:', mean_absolute_error(y_train, pred))
    print()
```

```python
#Preformance on the dev set:
for name, model in fitted_models.items():
    pred = model.predict(X_dev)
    print( name )
    print( '--------' )
    print( 'R^2:', r2_score(y_dev, pred ))
    print( 'MAE:', mean_absolute_error(y_dev, pred))
    print()
```

Gradient boosting model and random forest seem to outperform other models. However, both GB and RF models suffer from overfitting. GB seems to perform a little bit better than RF, so I am going to focus on tuning a GB model. 

## Tuning Gradient boosting model

```python
fitted_models['gb'].best_estimator_
```

```python
#Plot importance of features for gradient boosting model:
GBmodel = GradientBoostingRegressor(learning_rate=0.05, n_estimators=700,max_depth=7,random_state=123)
GBmodel.fit(X_train,y_train)
```

```python
predictors=list(X_train)
feat_imp = pd.Series(GBmodel.feature_importances_, predictors).sort_values(ascending=False)
feat_imp=feat_imp[:25]
feat_imp.plot(kind='bar', title='Importance of Features')
plt.ylabel('Feature Importance Score')
```

I am going to start with tuning the 'boosting' parameters: number of learners and learning rate. I am going to run a randomized search cross-validation this time over a larger set of hyperparameters. 

```python
gb_hyperparameters = { 
    'gradientboostingregressor__n_estimators': range(400,1601,200),
    'gradientboostingregressor__learning_rate' : [0.01,0.05,0.1],
    'gradientboostingregressor__min_samples_split': [2,10,50,100,200,500],
    'gradientboostingregressor__min_samples_leaf': [5,10,15,20,30,40,50],
    'gradientboostingregressor__max_depth': [4,5,6,7,8],
    'gradientboostingregressor__max_features': range(7,35,5),
    'gradientboostingregressor__subsample': [0.75,0.8,0.85,0.9]
}

hyperparameters['gb']=gb_hyperparameters

name='gb'

model = RandomizedSearchCV(pipelines[name], param_distributions=hyperparameters[name], n_iter=200, cv = 3, verbose=1, n_jobs = -1,random_state=123)

# Fit model on X_train, y_train
model.fit(X_train, y_train)

# Store model in fitted_models[name] 
fitted_models[name] = model

#Performance on the train set:
pred = model.predict(X_train)
print( name )
print( '--------' )
print('Performance on the train set:')
print( 'R^2:', r2_score(y_train, pred ))
print( 'MAE:', mean_absolute_error(y_train, pred))
print()

#Performance on the dev set:
pred = model.predict(X_dev)
print( '--------' )
print('Performance on the dev set:')
print( 'R^2:', r2_score(y_dev, pred ))
print( 'MAE:', mean_absolute_error(y_dev, pred))
print()
```

MAE decreased quite a bit! Success. 

```python
fitted_models['gb'].best_params_
```

```python
#Performance on the train set:
pred = GBmodel.predict(X_train)
print( name )
print( '--------' )
print('Performance on the train set:')
print( 'R^2:', r2_score(y_train, pred ))
print( 'MAE:', mean_absolute_error(y_train, pred))
print()

#Performance on the dev set:
pred = GBmodel.predict(X_dev)
print( '--------' )
print('Performance on the dev set:')
print( 'R^2:', r2_score(y_dev, pred ))
print( 'MAE:', mean_absolute_error(y_dev, pred))
print()
```

Final attemp to tune the hyperparameters a bit better:

```python
gb_hyperparameters = { 
    'gradientboostingregressor__n_estimators': [7000,7500],
    'gradientboostingregressor__learning_rate' : [0.01],
    'gradientboostingregressor__min_samples_split': [100],
    'gradientboostingregressor__min_samples_leaf': [15],
    'gradientboostingregressor__max_depth': [8,9,10],
    'gradientboostingregressor__max_features': [17],
    'gradientboostingregressor__subsample': [0.9]
}

hyperparameters['gb']=gb_hyperparameters

name='gb'

model =  GridSearchCV(pipelines[name], hyperparameters[name], cv = 3, verbose=0, n_jobs = -1)

# Fit model on X_train, y_train
model.fit(X_train, y_train)

# Store model in fitted_models[name] 
fitted_models[name] = model

#Performance on the train set:
pred = model.predict(X_train)
print( name )
print( '--------' )
print('Performance on the train set:')
print( 'R^2:', r2_score(y_train, pred ))
print( 'MAE:', mean_absolute_error(y_train, pred))
print()

#Performance on the dev set:
pred = model.predict(X_dev)
print( '--------' )
print('Performance on the dev set:')
print( 'R^2:', r2_score(y_dev, pred ))
print( 'MAE:', mean_absolute_error(y_dev, pred))
print()
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
for name, model in fitted_models.items():
    pred = model.predict(X_test)
    print( name )
    print( '--------' )
    print( 'R^2:', r2_score(y_test, pred ))
    print( 'MAE:', mean_absolute_error(y_test, pred))
    print( 'MAPE:', mean_absolute_percentage_error(y_test, pred))
    print()
```
